# Step 1: Library Imports
#### Basic imports for image processing and clustering

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Step 2: Image Loading and Display
#### Loads and displays original image

```python
# Reads the image file "flower.jpg" using matplotlib
image = mpl.image.imread("flower.jpg")
```

```python
# Displays the image using plt.imshow()
plt.imshow(image)
```

```python
# image.shape shows the dimensions of the image (height × width × 3 color channels)
image.shape
```

# Step 3: Image Preprocessing
#### Reshapes image for clustering

```python
X = image.reshape(-1,3)
# Reshapes the 3D image array into a 2D array
```

```python
X.shape
# Below Each row represents one pixel
# 3 columns represent RGB values
```

# Step 4: K-Means Clustering
#### Applying Clustering Algorithm

```python
kmeans = KMeans(n_clusters= 5,n_init = 10)
# Creates K-Means clustering object with 10 clusters
# n_init=10 means algorithm runs 10 times with different initializations
```

```python
kmeans.fit(X)
# Fits the model to the reshaped image data
```

# Step 5: Image Segmentation
#### Creates segmented image

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)

"""
kmeans.labels_ assigns each pixel to a cluster
kmeans.cluster_centers_ contains the average color of each cluster
Replaces each pixel with its corresponding cluster center color
Reshapes back to original image dimensions
"""
```

# Step 6: Display Segmented Image

```python
plt.imshow(segmented_img/255)
# Displays the segmented image
# Divides by 255 to normalize pixel values for proper display
```

# Step 7: Save Results

```python
import cv2

cv2.imwrite("im1.jpg",cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
cv2.imwrite("im2.jpg",cv2.cvtColor(segmented_img.astype("uint8"),cv2.COLOR_BGR2RGB))

# imports OpenCV library
# Saves original image as "im1.jpg"
# Saves segmented image as "im2.jpg"
# Converts color space from BGR to RGB (OpenCV uses BGR by default)
# astype("uint8") converts segmented image to 8-bit unsigned integer format
```
