# Importing important Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

# Loading the Dataset

```python
data = pd.read_csv("predictive_maintenance.csv")
```

```python
data.head()
```

# Dataset Analysis

```python
data.info()
```

```python
sns.countplot(y = "Failure Type", data = data)
plt.show()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.drop('UDI', axis=1) #remove unwanted columns
```

# Label Encode Categorical Columns


```python
label = LabelEncoder()
```

```python
data[['Product ID','Type','Failure Type']] = data[['Product ID','Type','Failure Type']].apply(label.fit_transform)
```

```python
data.head(5)
```

# Data Visualization

```python
sns.pairplot(data)
plt.show()
```

```python
correlation = data.corr()
plt.figure(figsize =  (20, 10))
sns.heatmap(correlation, annot = True, annot_kws={"size" : 12})
plt.title("Correlation Heatmap")
plt.tick_params(labelsize = 8)
plt.show()           
           
```

```python
# Checkiing Relationship Between Independent and Dependent
for col in data[:-1]:
    plt.scatter(x = data[col], y = data["Failure Type"])
    plt.title(f"Relationship Between {col} and Failure Type")
    plt.show()
```

# load Tools 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import  RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x = data.iloc[:,:-1]
y = data["Failure Type"]
```

```python
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 9)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# fitting data to model
model = RandomForestClassifier(n_estimators = 100, random_state = 9, criterion = "gini", class_weight = "balanced")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
evaluation = pd.DataFrame({
    "Actual Value" : y_test,
    "Predicted Value" : y_pred,
    "Error" : y_test - y_pred
})
evaluation
```

```python
# Checking accuracy 
accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Model Accuracy is {accuracy}")
```

```python
# confusion matrix
confusion_m = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(confusion_m, annot = True, fmt = "d", color = "blue")
plt.title("Confusion Metrix Heatmap")
plt.show()
```

```python
# Save the model
import joblib
joblib.dump(model, "Maintainance_Predictor.joblib")

```
